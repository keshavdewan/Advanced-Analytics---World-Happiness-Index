```python
# Creating a subset excluding 'Country name', 'Country Code', 'Regional Indicator', 'year'

df_sub = df_whr[['Happiness Score', 'Log GDP per capita', 'Social support',
       'Healthy life expectancy at birth', 'Freedom to make life choices',
       'Generosity', 'Perceptions of corruption']]
```

```python
df_sub
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_sub.corr(), annot = True, ax = ax) 
# The 'annot' argument allows the plot to place the correlation coefficients onto the heatmap.
```

###### Correlation findings :
 Social Support, 0.68 - Strong correlation - This means a country with higher social support no.s will have higher happiness score
 
Perception of Corruption, -0.31 - This means lower is the perception of corruption, more is the happiness score in a country

Other factors doesn't seem to have strong effect to the happiness score

### Categorical Plot

Creating a categorical variable that splits 'Happiness Score' column into categories

```python
# Using a histogram to visualize the distribution of the variables.

sns.histplot(df_whr['Happiness Score'], bins = 25, kde = True)
```

```python
df_whr.loc[df_whr['Happiness Score'] < 4, 'Happiness Category'] = 'Less Happy'
```

```python
df_whr.loc[(df_whr['Happiness Score'] >= 4) & (df_whr['Happiness Score'] < 7), 'Happiness Category'] = 'Moderately Happy'
```

```python
df_whr.loc[df_whr['Happiness Score'] >= 7, 'Happiness Category'] = 'Happy'
```

```python
df_whr['Happiness Category'].value_counts(dropna = False)
```

```python
df_whr.columns
```

```python
%%time
# Creating a categorical plot in seaborn using the 'Happiness categories'

sns.set(style = 'ticks')
g = sns.catplot(x = 'Freedom to make life choices', y = 'Happiness Score', hue = 'Happiness Category', data = df_whr)
```

From this graph, we can see that the countries that are happy have a higher 'freedom to make life choices' number (as per the clutter), while at very few instances the 'freedom to make life choices' is low and the country happiness score is high there.
But, this graph doesn't give the clear picture as many countries have higher freedom to make life choices but are also 'moderately happy' and 'less happy'

```python
%%time
# Creating a categorical plot in seaborn using the 'Happiness categories'

sns.set(style = 'ticks')
q = sns.catplot(x = 'Perceptions of corruption', y = 'Happiness Score', hue = 'Happiness Category', data = df_whr)
```

With this categorical plot,  we can see a green cluster above, this shows that countries with less perception of corruption score are more happy

```python
%%time
# Creating a categorical plot in seaborn using the 'Happiness categories'

sns.set(style = 'ticks')
w = sns.catplot(x = 'Generosity', y = 'Happiness Score', hue = 'Happiness Category', data = df_whr)
```

Hypothesis - Apart from all the factors, Social Security and Perceptions of Corruption have a huge impact on a country's Happiness Score
